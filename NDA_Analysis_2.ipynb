```python
import pandas as pd #Data Frames
import numpy as np #Statistics
import matplotlib.pyplot as plt #plots
import matplotlib as mpl #plots

data = pd.read_csv('') #import csv to pandas
```

```python
pd.set_option('display.float_format', '{:20,.3f}'.format)
data.describe() #statistical summary of data frame's columns
```

```python
CE = {}
cc = 0
for i in data['RAZON SOCIAL']:
  if i not in CE:
    CE.update({i:data['EMPRESA MEXICO'][cc]}) #Create a dict with the name of each individual importer and their RFC ID
    # print(i+': ',CE[i])
  cc += 1

dfCE = pd.DataFrame({'Importador':list(CE.keys()), 'RFC':list(CE.values())}) #Transform dict into df and choose column names
# dfCE.to_excel('Importadores_RFC.xlsx') #import df into excel file
```

```python
#Let´s create a dictionary with the customs name as key and an individual colour value as the value
rgb = [] #Create color list
for name, hex in mpl.colors.cnames.items():
    rgb.append(hex) #Add hex color codes to list

aduanas = data['ADUANA'].unique().tolist() #Send the names on the column into a list
colorcodeA = {}
for i in range(0,len(aduanas)):
  colorcodeA.update({aduanas[i]: rgb[i+10]})


importadores = data['RAZON SOCIAL'].unique().tolist() #Send the names on the column into a list
colorcodeI = {}
for i in range(0,len(importadores)):
  colorcodeI.update({importadores[i]: rgb[i+10]})
```

```python
fig, ax = plt.subplots() #Create figure
for A in aduanas: #We use the prior customs list to create plots by each custom
  ax.scatter('MMBtu', 'USD', c = colorcodeA[A], label = A.split(',')[0], edgecolor='k', alpha = 0.5, data = data[data['ADUANA'] == A]) # scatter plot by custom
#) & (data['MMBtu'] <= 1*1e8) & (data['USD'] <= 3*1e7) #in case we want to filter the MMBtus or USDs
ax.set_xlabel('MMBtu')
ax.set_ylabel('USD')
# ax.set_yscale('log')
# ax.set_xscale('log')
plt.legend(aduanas, loc='center right', bbox_to_anchor=(1.85, 0.5), fontsize = 7)
plt.title('MMBtu vs Precio por cada Aduana')
plt.grid(linestyle = '--', alpha = 0.5)
plt.show()
```

```python
# @title
cc = 0
NI = []
for I in importadores:
  if I not in ['CFENERGIA SA DE CV', 'TECHGEN SA DE CV']: #This option filters specific values. To dispose, just write rubish or add a character to the options.
    if (cc < 30): #Filter the number of values to evaluate
      NI.append(I) #Make sure the legend matches the actual values analized
      avg_MMBtu = data[data['RAZON SOCIAL'] == I].groupby('MES')['MMBtu'].agg('mean')
      avg_MMBtu.plot(title = 'Promedio de MMBtu por Mes', xlabel = 'Mes', ylabel = 'MMBtu')
  cc += 1
plt.legend(NI, loc='center right', bbox_to_anchor=(1.6, 0.5), fontsize = 6)
# plt.yscale('log')
plt.grid(axis = 'x', linestyle = '--', alpha = 0.5)
plt.show()

data.groupby('MES')['MMBtu'].agg('mean').plot(title = 'Promedio de MMBtu por Mes', xlabel = 'Mes', ylabel = 'MMBtu')
plt.grid(axis = 'x', linestyle = '--', alpha = 0.5)
plt.show()
```

```python
cc = 0
NI = []
for I in importadores:
  if I not in ['CFENERGIA SA DE CV', '+PEMEX TRANSFORMACION INDUSTRIAL EPS', '+IENOVA MARKETING S DE RL DE CV']: #This option filters specific values. To dispose, just write rubish or add a character to the options.
    if (cc < 300): #Filter the number of values to evaluate
      NI.append(I) #Make sure the legend matches the actual values analized
      sum_MMBtu = data[data['RAZON SOCIAL'] == I].groupby('MES')['MMBtu'].agg('sum')
      sum_MMBtu.plot(title = 'Suma de MMBtu por Importador a lo Largo del Tiempo', xlabel = 'Mes', ylabel = 'MMBtu')
  cc += 1
plt.legend(NI, loc='center right', bbox_to_anchor=(1.6, 0.5), fontsize = 6)
# plt.yscale('log')
plt.grid(axis = 'x', linestyle = '--', alpha = 0.5)
plt.show()

data.groupby('MES')['MMBtu'].agg('sum').plot(title = 'Suma de MMBtu por Mes', xlabel = 'Mes', ylabel = 'MMBtu')
plt.grid(axis = 'x', linestyle = '--', alpha = 0.5)
plt.show()
```

```python
RS = dict(data.groupby('RAZON SOCIAL')['MMBtu'].agg('mean'))
RS = dict(sorted(RS.items(), key = lambda x:x[1]))
plt.barh(list(RS.keys()), list(RS.values()))
plt.yticks(fontsize = 6, alpha = 0.8);
plt.xlabel('MMbtu (Primedio)')
plt.ylabel('Importador')
plt.title('Promedio de MMBtu por Importador');
plt.grid(axis = 'x', linestyle = '--', alpha = 0.5)
```
